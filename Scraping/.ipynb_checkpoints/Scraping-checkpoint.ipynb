# Partie scraping du projet DSIA Data engineering 

```python
import requests
import string
from bs4 import BeautifulSoup
```

```python
mot_cle = "informatique"
page = 0

url = "".join(["https://www.indeed.fr/emplois?q=", mot_cle, "&start=", str(page)])

response = requests.get(url)
soup = BeautifulSoup(response.text)
```

```python
soup
```

```python
soup.find_all(class_='jobtitle')
```
